```python
import json
import os
import pandas as pd

json_path = r"C:\research_projects\video_analysis_pipeline_research\results\processing_results.json"

with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

videos = data["results"]

print(f"Videos in results: {len(videos)}")
```

```python
records = []

for v in videos:
    extraction = v.get("extraction")

    has_extraction = extraction is not None
    has_error = False
    error_message = None
    error_code = None

    if isinstance(extraction, dict) and "error" in extraction:
        has_error = True
        error_message = extraction.get("error")

        for code in ["429", "504", "503", "500", "403", "401"]:
            if error_message and code in error_message:
                error_code = code
                break

    records.append({
        "video_name": v.get("video_name"),
        "has_extraction": has_extraction,
        "has_error": has_error,
        "error_code": error_code,
    })

df = pd.DataFrame(records)
df.head()
```

```python
records = []

for v in videos:
    extraction = v.get("extraction")

    error_message = None
    error_code = None

    if isinstance(extraction, dict) and "error" in extraction:
        error_message = extraction["error"]
        for code in ["429", "504", "503", "500", "403"]:
            if code in error_message:
                error_code = code
                break

    if error_code:
        records.append({
            "video_name": v.get("video_name"),
            "video_path": v.get("video_path"),
            "error_code": error_code,
            "error_message": error_message,
            "processing_time_s": v.get("pipeline_stats", {}).get("processing_time_s"),
            "final_frame_count": v.get("pipeline_stats", {}).get("final_frame_count"),
        })

error_df = pd.DataFrame(records)
# error_df.to_csv("videos_with_extraction_errors_detailed.csv", index=False)
error_df
```

```python
# Filter videos that have extractions (excluding None and error-only extractions)
filtered_videos = []
removed_videos = []

for v in videos:
    extraction = v.get("extraction")
    has_valid_extraction = extraction is not None and not (isinstance(extraction, dict) and "error" in extraction and len(extraction) == 1)
    
    if has_valid_extraction:
        filtered_videos.append(v)
    else:
        removed_videos.append(v.get("video_name"))

print(f"Filtered video count: {len(filtered_videos)}")
print(f"Removed video count: {len(removed_videos)}")

# Create new filtered results file
filtered_data = {
    "metadata": {
        "total_videos": len(filtered_videos),
        "original_total_videos": data["metadata"]["total_videos"],
        "videos_removed": len(removed_videos),
        "processing_date": data["metadata"].get("processing_date"),
    },
    "results": filtered_videos
}

# Save filtered results
output_path = r"C:\research_projects\video_analysis_pipeline_research\results\processing_results_filtered.json"
with open(output_path, "w") as f:
    json.dump(filtered_data, f, indent=2)

print(f"\nFiltered results saved to: {output_path}")

# Save list of removed videos
removed_df = pd.DataFrame({"video_name": removed_videos})
removed_csv_path = r"C:\research_projects\video_analysis_pipeline_research\results\removed_videos.csv"
removed_df.to_csv(removed_csv_path, index=False)
print(f"List of removed videos saved to: {removed_csv_path}")

# Print summary statistics
print("\n--- SUMMARY ---")
print(f"Original videos: {data['metadata']['total_videos']}")
print(f"Videos with valid extractions: {len(filtered_videos)}")
print(f"Videos removed (no extraction): {len(removed_videos)}")
print(f"Retention rate: {len(filtered_videos)/data['metadata']['total_videos']*100:.1f}%")
```

```python
# Verify what's being removed
print("Checking removed videos...")

removed_details = []
for v in videos:
    extraction = v.get("extraction")
    video_name = v.get("video_name")
    
    if video_name in removed_videos:
        removed_details.append({
            "video_name": video_name,
            "extraction_is_none": extraction is None,
            "extraction_type": type(extraction).__name__ if extraction is not None else "None",
            "has_error_key": "error" in extraction if isinstance(extraction, dict) else False,
            "extraction_keys": list(extraction.keys()) if isinstance(extraction, dict) else None,
            "extraction_preview": str(extraction)[:200] if extraction else None
        })

removed_check_df = pd.DataFrame(removed_details)
print(f"\nTotal removed: {len(removed_check_df)}")
print(f"Where extraction is None: {removed_check_df['extraction_is_none'].sum()}")
print(f"Where extraction has error key: {removed_check_df['has_error_key'].sum()}")

# Show some examples
print("\n--- Sample of removed videos ---")
print(removed_check_df.head(10))

# Save detailed report
removed_check_df.to_csv(r"C:\research_projects\video_analysis_pipeline_research\results\removed_videos_detailed.csv", index=False)
print("\nDetailed report saved to: removed_videos_detailed.csv")
```
